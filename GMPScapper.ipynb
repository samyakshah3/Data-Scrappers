```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

def driver_define():
    print('Chromedriver Installing')
    driver_path = ChromeDriverManager().install()
    
    print('Chrome Browser Opening')
    options = Options()
    options.add_experimental_option('excludeSwitches', ['enable-logging'])
    
    driver = webdriver.Chrome(service=Service(driver_path), options=options)
    
    return driver

# Define a filename for the CSV file
filename = "output_data.csv"

e = []

def extract_website_links(browser, link):
    browser.get(link)
    time.sleep(10)  # pausing the program for 10 seconds

    source = browser.page_source

    # Beautiful Soup for scraping the HTML source
    soup = BeautifulSoup(source, 'html.parser')
    
    website_links = []
    for link_element in soup.find_all('a', {"class": "yYlJEf Q7PwXb L48Cpd brKmxb"}):
        website = link_element.get('href')
        website_links.append(website)

    return website_links

def Selenium_extractor(browser, page_number):
    time.sleep(2)

    # Extract the website links inside the loop to avoid stale elements
    website_links = extract_website_links(browser, browser.current_url)

    a = browser.find_elements(By.CSS_SELECTOR, ".cXedhc")
    time.sleep(1)

    records = []  # Create a list to store records
    for i in range(len(a)):
        a[i].click()
        time.sleep(2)

        # Re-fetch the source inside the loop
        source = browser.page_source

        # Beautiful Soup for scraping the HTML source
        soup = BeautifulSoup(source, 'html.parser')
        try:
            Name_Html = soup.findAll('div', {"class": "SPZz6b"})
            name = Name_Html[0].text if Name_Html else "Name not available"
            Phone_Html = soup.findAll('span', {"class": "LrzXr zdqRlf kno-fv"})
            phone = Phone_Html[0].text if Phone_Html else "Phone not available"
            Address_Html = soup.findAll('span', {"class": "LrzXr"})
            address = Address_Html[0].text if Address_Html else "Address not available"

            # Include website from the extracted links
            if i < len(website_links):
                website = website_links[i]
            else:
                website = "Website not available"

            # Include website in the record along with page number
            record = (page_number, website, name, phone, address)
            records.append(record)

        except:
            print("error")
            continue

    # Write all records to the CSV file
    df = pd.DataFrame(records, columns=['Page Number', 'Website', 'Name', 'Phone number', 'Address'])
    df.to_csv(filename, mode='a', header=False, index=False, encoding='utf-8')

def click_next_button(browser):
    try:
        # Locate the "Next" button using a CSS selector
        next_button = browser.find_element(By.CSS_SELECTOR, "a#pnnext")
        next_button.click()
    except Exception as e:
        print("ERROR occurred at clicking the next button:", e)

# Input URL from the user
print("Enter the link of the page")
link = input()
browser = driver_define()  # Initialize the driver

# Create a new CSV file with headers
initial_df = pd.DataFrame(columns=['Page Number', 'Website', 'Name', 'Phone number', 'Address'])
initial_df.to_csv(filename, index=False, encoding='utf-8')

# Visit the Google search results page
browser.get(link)

# Inside the main loop, you can keep track of the page number
page_number = 1
while True:
    Selenium_extractor(browser, page_number)
    try:
        click_next_button(browser)
        page_number += 1  # Increment page number after scraping a page
    except:
        break
```
